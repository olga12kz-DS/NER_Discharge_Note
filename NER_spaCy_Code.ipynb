# **TEXT EXTRACTION AND CLEAN-UP**

From the full record, extract the "History of Present Illness" column which is located, generally, between "Allergies" and

```python
# User Defined Function to extract History of Present Illness: (everything after "History of Present Illness:" until break) and add it as a column to the original dataframe.

def extract_present_illness(pattern, df, column_name="text"):
    """
    Extracts matches based on a pre-defined regex pattern from a specified column in the dataframe.

    Parameters:
        pattern (str): The regular expression pattern to use for extraction: r'History of Present Illness:\s*(.*)'
        df (pd.DataFrame): The DataFrame containing the text column: df_discharge_text.
        column_name (str): The name of the column to search, "text".

    Returns:
        pd.Series: A Series containing the extracted matches which will be added as a new column to the original dataframe.
    """
    # applying the pre-defined regex to each row of the target column
    return df[column_name].apply(lambda text: re.search(pattern, str(text)).group(1) if re.search(pattern, str(text)) else None)


# Define regex pattern
pattern = pattern = r'(?s)History of Present Illness:\s*(.+?)(Past Medical History:|\Z)'

# \n\s*\n stops at the first blank line (double newline) to prevent capturing unwanted sections.
# \Z ensures it captures everything until the end if no blank line is found.

# Apply function to extract service names
df_extract["Present Illness History Extracted"] = extract_present_illness(pattern, df_discharge_text, "text")
```

```python
# Since we captured all text as is for History of Present Illness, we need to remove unnecessary characters.
# Find and remove special characters and new lines.

df_extract["Present Illness Cleaned"] = df_extract["Present Illness History Extracted"].str.replace(r'(\s*\n\s*|_{3,}|={3,}|\+)', ' ', regex=True).str.strip()
df_extract.iloc[4]
```

Now the three models can be compared on performance, specifically

*   Number of identified entities
*   Type of indentified entities
*   Processing time/speed


```python
# User Defined Function to measure execution time and entity extraction performance.
def evaluate_model(nlp, text): # takes the model and the text to be processed by the model
    start_time = time.time() # record start time
    doc = nlp(text) # process the text by each model
    elapsed_time = time.time() - start_time # total processing time for the "doc"

    entities = [(ent.text.lower(), ent.label_) for ent in doc.ents] #ent.text extracts entity text (case-insensitive), ent.label_ extracts NER label from the "doc"
    entity_count = len(entities) # counts the number of extracted entities
    entity_types = set([ent[1] for ent in entities]) # takes the list of entities, creates a unique set of the second element of the entity tuple,
    # Example of entity tuple: [("COVID-19", "DISEASE"), ("Remdesivir", "CHEMICAL")], i.e., list with immutable pairs.
    return elapsed_time, entity_count, entity_types, entities # return the metrics

# Run evaluations
results = { # apply the function to each model and store results as a dictionary where the key is the model and the value is the output of the function.
    "en_core_sci_sm": evaluate_model(nlp_sm, sample),
    "en_core_sci_md": evaluate_model(nlp_md, sample),
    "en_ner_bc5cdr_md": evaluate_model(nlp_bc, sample),
}

# Print results
for model, (time_taken, entity_count, entity_types, entities) in results.items(): # from each item in the results dictionary, print the model and metrics.
    print(f"Model: {model}")
    print(f"Processing Time: {time_taken:.4f} seconds")
    print(f"Total Entities Detected: {entity_count}")
    print(f"Entity Types: {entity_types}")
    print(f"Entities: {entities}\n")
```

```python
# Compare unique extracted entities (text) among the three models.
def compare_entities(nlp, text):
    doc = nlp(text)

    entities = [(ent.text, ent.label_) for ent in doc.ents]
    text_entities_unique = set([ent[0].lower() for ent in entities]) # .lower to render the text case-insensitive
    return text_entities_unique


# Run evaluations
results = {
    "en_core_sci_sm": compare_entities(nlp_sm, sample),
    "en_core_sci_md": compare_entities(nlp_md, sample),
    "en_ner_bc5cdr_md": compare_entities(nlp_bc, sample),
}

# Print results
for model, text_entities_unique in results.items(): # ensure text_entities_unique is still a set (unordered and mutable but unique) and not a tuple (ordered and immutable but duplicate-permissible)
    print(f"Model: {model}")
    print(f"Unique Entities: {text_entities_unique}")
    # print the number of unique entities in each model
    print(f"Numer of unique entities in the model: {len(text_entities_unique)}")
```

```python
# Find entities detected by all thee three models (intersection, i.e., shared entities).
common_all = results["en_core_sci_sm"] & results["en_core_sci_md"] & results["en_ner_bc5cdr_md"]
print("Entities common to all models:", common_all)
len(common_all)
```

```python
# Find entities unique to each model (union).
print("Unique to sm:", results["en_core_sci_sm"] - results["en_core_sci_md"] - results["en_ner_bc5cdr_md"])
print("Unique to md:", results["en_core_sci_md"] - results["en_core_sci_sm"] - results["en_ner_bc5cdr_md"])
print("Unique to bc:", results["en_ner_bc5cdr_md"] - results["en_core_sci_sm"] - results["en_core_sci_md"])
```

# **IMPROVING BASELINE PERFORMANCE**

The baseline performance can be improved by combining regex rules to extract drugs/dose/route/frequency.

```python
# Define the extraction rule which must ignore punctuation, extra spaces, and case-insensitive.
# To ensure detection of only drug names, the presence of dose is required.
pattern = re.compile(
    r"\b([a-zA-Z][a-zA-Z0-9\-]*)"                              # drug name
    r"\s+(\d+(?:\.\d+)?)(mg|g|mcg|units)"                      # required dose + unit
    r"(?:\s+(IV|PO|IM|SC|PR))?"                                # optional route
    r"(?:\s*x\s*(\d+))?",                                      # optional frequency
    re.IGNORECASE
)
```

```python
matches = pattern.findall(sample)

results = []
for match in matches:
    drug, dose, unit, route, freq = match
    result = {
        "drug": drug.strip(),
        "dose": (dose + unit) if dose else None,
        "route": route,
        "frequency": f"x{freq}" if freq else None
    }
    results.append(result)

# Display structured output
pd.DataFrame(results)
```

We will now proceed to testing medspacy with defined extraction rules.

```python
# Unlike spaCy, which requires downloading a specific pre-trained model, medspaCy builds its pipeline by wrapping a base spaCy model. If required, however,
# wrapping can be done around a domain-specific model like one of the three we have already tested. We will start with a base model.
med_model = medspacy.load() # This will download the pipeline which contains TargetMatcher which we need for pattern-matching and extraction.
```

```python
# Create a variable to hold TargetMatcher object.
matcher = med_model.get_pipe("medspacy_target_matcher") # retrieve the TargetMatcher component from the pipeline by using its registered name.
```

```python
# Create a simple rule from "sample" to test.
rule = TargetRule("morphine", "MEDICATION") # define entity text to detect and a corresponding entity label.
matcher.add([rule]) # add the rule to the TargetMatcher object
```

```python
# Process the text and print the result.
doc = med_model(sample)
print([(ent.text, ent.label_) for ent in doc.ents])
```

Pre-trained models in spaCy can extract entity text and labels without pre-defined rules as they use a general statistical approach in NER. medspaCy, however, while more precise, requires pre-defined rules.

```python
# We can combine regex and medspaCy's TargetMatcher to process the text through a flexible pattern.
matcher2 = med_model.get_pipe("medspacy_target_matcher")

# Define regex-based rule. It should capture drugs/doses/route/frequency (everything after "drug" is optional), ignoring punctuation and extra spaces. The patter is case-insensitive.
rule2 = TargetRule(
    r"\b[a-zA-Z][a-zA-Z0-9\-]{2,}\b(?:\s*\d+(?:mg|g|mcg|units)?)?(?:\s*(IV|PO|IM|SC|PR))?(?:\s*x\d+)?",
    "MEDICATION", pattern = "regex"
)

matcher2.add([rule2])

# Run the NLP pipeline
doc2 = med_model(sample)

# Print extracted medication mentions
for ent in doc2.ents:
    print(ent.text.strip(), "", ent.label_)
```
